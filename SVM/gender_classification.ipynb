```python
import pandas as pd
```

```python
df=pd.read_csv('./gender_classification_v7.csv')
df
```

```python
target=pd.get_dummies(df['gender'])
target=list(target.loc[:,'Male'])
target
```

```python
df.insert(8,'target',target)
```

```python
df
```

```python
x=df.iloc[:,0:7]
x
```

```python
y=df['target']
y
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=SVC(kernel='linear',C=2,gamma=10,probability=True)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
import numpy as np
```

```python
test_df=pd.DataFrame(np.c_[y_pred,y_test],columns=['predict','actual'])
test_df
```

```python
x_test
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
x_train2,x_test2,y_train2,y_test2=train_test_split(x,y,test_size=0.2)
```

```python
model1=SVC(kernel='rbf',C=2.5,gamma=5.5,probability=True)
model1.fit(x_train2,y_train2)
```

```python
y_pred2= model1.predict(x_test2)
print(accuracy_score(y_test2,y_pred2))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
pred_prob1=model.predict_proba(x_test)
pred_prob2=model.predict_proba(x_test2)
```

```python
pred_prob1.shape
```

```python
fpr,tpr,thresh=roc_curve(y_test,pred_prob1[:,1],pos_label=1)
fpr1,tpr1,thresh1=roc_curve(y_test2,pred_prob2[:,1],pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score1 = roc_auc_score(y_test, pred_prob1[:,1])
print(auc_score1)
auc_score2 = roc_auc_score(y_test2, pred_prob2[:,1])
print(auc_score2)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
plt.plot(fpr,tpr,linestyle='dashed',color='orange',label="model-1")
plt.plot(fpr1,tpr1,linestyle='dashed',color='green',label="model-2")
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.show()
```
